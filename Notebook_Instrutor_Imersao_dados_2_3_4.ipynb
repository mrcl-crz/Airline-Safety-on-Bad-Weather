# **Aula 3 - Tratando a base com os principais comandos de Pandas**


```python
#Descartando colunas que não vamos usar

df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()

df_principal
```

```python
#Renomeando colunas

df_principal = df_principal.rename(columns = {'Último (R$)': 'Valor_final', 'Var. Dia (%)': 'Var_dia_pct'}).copy()

df_principal
```

```python
#criando as demais colunas - variacao_pct e valor_inicial

df_principal['Var_pct'] = df_principal['Var_dia_pct']/100


df_principal['valor_inicial'] = df_principal['Valor_final']/(1+df_principal['Var_pct'])


df_principal.head(10)
```

```python
df_principal = df_principal[['Ativo', 'ETC']].copy()
```

```python
#criando as demais colunas - quantidade de acoes

df_principal = df_principal.merge(df_total_acoes, left_on='Ativo', right_on='Código', how='left')

df_principal = df_principal.drop(columns=['Código']) #fazer sem essa linha e depois incluir

df_principal.head(10)
```

```python
#criando as demais colunas - variacao_$

df_principal['Variacao_RS'] = (df_principal['Valor_final']- df_principal['valor_inicial'])*df_principal['Qtde. Teórica']


#alterando o formato que aparece os valores

pd.options.display.float_format = '{:.2f}'.format

#convertendo quantidade teórica para inteiro

df_principal['Qtde. Teórica'] = df_principal['Qtde. Teórica'].astype(int)


#renomeando a coluna de qtde teórica

df_principal = df_principal.rename(columns = {'Qtde. Teórica': 'Qtde_Teorica'}).copy()


df_principal.head(10)
```

```python
#criando as demais colunas - Resultado

#PERGUNTAR PARA O CHATGPT COMO REPRODUZIR A FORMULA, ELE VAI TRAZER DE UM JEITO LONGO MAS PODE PEDIR PRA FAZER EM UMA LINHA COM FUNCAO LAMBDA


df_principal['Resultado'] = df_principal['Variacao_RS'].apply(lambda x: 'Estável' if x == 0 else ('Subiu' if x > 0 else 'Desceu'))


df_principal.head(10)
```

```python
#criando as demais colunas - nome, segmento, idade e cat_idade

#nome da empresa

df_principal = df_principal.merge(df_ticker, left_on='Ativo', right_on='Ticker', how='left')

df_principal = df_principal.drop(columns=['Ticker'])

#segmento

df_principal = df_principal.merge(df_gpt[['Nome da empresa','Segmento','Idade']], left_on='Nome empresa', right_on='Nome da empresa', how='left')

df_principal = df_principal.drop(columns=['Nome empresa'])

#cat_idade

df_principal['Cat_idade'] = df_principal['Idade'].apply(lambda x: "Mais de 100 anos" if x > 100 else ("Entre 100 e 50 anos" if x >= 50 else "Menos que 50 anos"))


df_principal.head(10)
```

# **Aula 3.2 - Gerando análises e gráficos**

```python
#Análise GERAL

#perguntar para o Chatgpt: Tenho uma planilha com aa colunas Variacao_rs e Resultado (subiu ou desceu). a partir dessas colunas monto a tabela abaixo com uso de fórmulas do excel. Se eu tivesse a mesma tabela em um dataframe do pandas como faria para chegar nos mesmos resultados?


# Calculando a maior variação
maior_variacao = df_principal['Variacao_RS'].max()

# Calculando a menor variação
menor_variacao = df_principal['Variacao_RS'].min()

# Calculando a média das variações
media_variacao = df_principal['Variacao_RS'].mean()

# Calculando a média das variações para os que subiram
media_variacao_subiram = df_principal[df_principal['Resultado'] == 'Subiu']['Variacao_RS'].mean()

# Calculando a média das variações para os que desceram
media_variacao_desceram = df_principal[df_principal['Resultado'] == 'Desceu']['Variacao_RS'].mean()

# Exibindo os resultados
print("Maior Variação: R$", maior_variacao)
print("Menor Variação: R$", menor_variacao)
print("Média das Variações: R$", media_variacao)
print("Média das Variações para os que subiram: R$", media_variacao_subiram)
print("Média das Variações para os que desceram: R$", media_variacao_desceram)


#DESAFIO, PRINTAR NA TELA O NOME DAS EMPRESAS COM MAIOR E MENOR VARIAÇÃO E FORMATAR OS VALORES
```
